```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df = df.set_index('PassengerId') #set passengerid as the new index 
df
```

```python
df.groupby('Sex').agg('count')['Survived']
```

```python
plt.title('')
plt.pie(x=np.array(df.groupby('Sex').agg('count')['Survived']),labels=['female','male'],autopct='%1.1f%%')
plt.legend(loc='upper left')
plt.title('Sex Pie')
```

```python
df['Sex'] = df['Sex'].map({'male':1,'female':0})
sns.pairplot(data=df[['Age','Sex','Fare']])
```

```python
sns.displot(data=df.Fare) 
plt.title('Fare Distribution')
```
