## 01. 筛选引擎

使用second_cars_info.csv制作一个查询引擎，可以通过输入指定的参数获取对应条目的结果。

```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
data = pd.read_csv('./exercise_data/second_cars_info.csv')
```

```python
# 数据预处理
data['New_price'] = data['New_price'].replace('(.*)万', '\\1', regex=True).replace('暂无',np.NaN).astype(float)
data['Km'] = data['Km'].replace('(.*)万公里', '\\1', regex=True).replace('百公里内',0.0).astype(float)
data['Sec_price'] = data['Sec_price'].astype(float)
data['Discharge'] = data['Discharge'].apply(lambda x:x.split(','))
data['Boarding_time'] = data['Boarding_time'].replace('([0-9]+)年([0-9]+)月', '\\1-\\2', regex=True).replace('未上牌', np.NaN)
data['Boarding_time'] = pd.to_datetime(data['Boarding_time'], format='%Y-%m')

display(data.head(20))
```

```python
props = data.columns.values.tolist()

print('以下条件可提供筛选：')
display(props)
```

### 开始逐步筛选

```python
# 对品牌筛选
brands = ['玛莎拉蒂', '保时捷']

result = data[data.Brand.isin(brands)] if not len(brands) == 0 else data
display(result)
```

```python
# 对名字进行关键字筛选
name = "3.0T"

result = result[result.Name.str.contains(name)]
display(result)
```

```python
from datetime import datetime

# 对时间进行范围筛选
date_start = datetime.strptime('1997-01', '%Y-%m')
date_end = datetime.strptime('2016-08', '%Y-%m')

result = result[(result['Boarding_time'] >= date_start) & (result['Boarding_time'] <= date_end)]
display(result)
```

```python
# 如果想找未上牌车辆
result = result[pd.isna(result['Boarding_time'])]

display(result)
```

```python
# 对里程进行范围筛选
km_range_low = 0
km_range_high = 5.0

result = result[result.Km.between(km_range_low, km_range_high)]
display(result)
```

```python
# 对排放标准进行筛选
disc = ['欧4']
disc = set(disc)
sfilter = disc.issubset

result = result[[sfilter(x) for x in result.Discharge.values.tolist()]]
display(result)
```

```python
# 对二手价格进行筛选
price_low = 0
price_high = 85

result = result[result.Sec_price.between(price_low, price_high)]
display(result)
```
