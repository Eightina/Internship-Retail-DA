```python
import pandas as pd
import numpy as np
```

```python
a = open('wind.txt', mode='r')
txtlist = a.readlines()
txtlist = np.array([i.split() for i in txtlist]).T
txtlist[0,1:] = ['19'+i for i in txtlist[0,1:]]
wind = {}
for o in txtlist:
    wind[o[0]] = o[1:]
    
wind = pd.DataFrame(wind)
wind
```

```python
temp = np.array(['-' for i in range(6574 )])
wind['DAT'] = np.array(wind['Yr'])+temp+np.array(wind['Mo'])+temp+np.array(wind['Dy'])
wind['DAT'] = pd.to_datetime(wind['DAT'])
wind.set_index('DAT',inplace=True)
wind = wind.astype("float")
wind
```

```python
wind.isna().sum()
```

```python
loc_stats = wind.drop(['Yr','Mo','Dy'],axis=1).describe(percentiles=[])
loc_stats = loc_stats.drop(['count','50%'],axis=0)
loc_stats
```

```python
day_stats = wind.drop(['Yr','Mo','Dy'],axis=1).T.describe(percentiles=[])
day_stats = day_stats.drop(['count','50%'],axis=0)
day_stats
```
