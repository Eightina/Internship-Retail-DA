```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('second_cars_info.csv')
df
```

```python
import re
keys = [o.split() for o in df.Name]
df['Type'] = [o[0] for o in keys]
df['Year'] = [re.search( r'20..', o) for o in df.Name]
df['Others'] = [o[2:] for o in keys]
for i in range(len(df)):
    try:
        df.Year[i] = int(df.Year[i].group())
    except AttributeError:
        df.Year[i] = np.nan
    except ValueError:
        df.Year[i] = np.nan
temp = [o.replace('年','-') for o in df['Boarding_time']] 
temp
```

```python
for (i,o) in enumerate(temp):
    try:
        df['Boarding_time'][i] = pd.to_datetime(o.replace('月',''))
    except ValueError:
        df['Boarding_time'][i] = pd.Timestamp(1970,1,1)
        
for (i,o) in enumerate(df['New_price']):
    try:
        df['New_price'][i] = float(o.replace('万',''))
    except ValueError:
        df['New_price'][i] = np.nan

for (i,o) in enumerate(df['Km']):
    try:
        df['Km'][i] = float(o.replace('万公里',''))
    except ValueError:
        df['Km'][i] = np.nan
df
```

```python
def search(brand=df.Brand, name=df.Name, year=df.Year, boarding=df.Boarding_time, km=df.Km, discharge=df.Discharge, sec=df.Sec_price, new=df.New_price, data=df): 
        Br = (brand==df.Brand)
        Na = (name==df.Name)
        Ye = (year==df.Year )
        y0,m0,y1,m1 = boarding
        Bo = ((df.Boarding_time >= [pd.Timestamp(y0,m0,1) for i in range(len(df))]) * (df.Boarding_time >= [pd.Timestamp(y1,m1,1) for i in range(len(df))]))
        Km = ((df.Km>=[km[0] for i in range(len(df))]) * (df.Km<=[km[1] for i in range(len(df))]))
        Di = np.array([(discharge[0] in i) for i in df.Discharge])
        Se = ((df.Sec_price>=[sec[0] for i in range(len(df))]) * (df.Sec_price<=[sec[1] for i in range(len(df))]))
        Ne = ((df.New_price>=[new[0] for i in range(len(df))]) * (df.New_price<=[new[1] for i in range(len(df))]))
        idx = Ye*Br*Na*Bo*Km*Di*Se*Ne
        result = df.loc[idx]
        return result
```

```python
search(brand='奥迪',
    year=2012,
    boarding=[2011,1,2012,8],
    km=[5,10],
    discharge=['OBD'],
    sec=[10,20],
    new=[0,100],
    data=df)
```
